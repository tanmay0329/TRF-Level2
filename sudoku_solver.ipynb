```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def fixColor(image):
    return (cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
img = cv2.imread("Sudoku_real.jpg")
plt.imshow(fixColor(img))
```

```python
print(img.shape)
height=img.shape[0]
width=img.shape[1]
print(height,width)


```

```python
img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(fixColor(img_gray))
```

```python


img_Blur = cv2.GaussianBlur (img_gray, (5,5),1)
plt.imshow(fixColor(img_Blur))
```

```python
img_Threshold = cv2.adaptiveThreshold(img_Blur,255,1,1,11,5)
plt.imshow(fixColor(img_Threshold))

```

```python
kernel=np.ones((3,3))
img_erode = cv2.erode(img_Threshold, kernel, iterations=1)
plt.imshow(fixColor(img_erode))
```

```python
img_dilute = cv2.dilate(img_erode, kernel, iterations=2)
plt.imshow(fixColor(img_dilute))
```

```python
img_contours = img.copy()

countours, _ = cv2.findContours(img_dilute,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
for cnt in countours:
    cv2.drawContours(img_contours,[cnt],0,(0,255,0),3)
plt.imshow(fixColor(img_contours))
```
